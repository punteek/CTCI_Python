```python
# 2.1 Remove Dupes: Write code to remove duplicates from an unsorted linked list
# Follow Up: How would you solve this problem if a temporary buffer is not allowed?

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeDupes(startNode: Node):
    prev = None

    seenSet = set()
    while(startNode != None):
        if startNode.data not in seenSet:
            seenSet.add(startNode.data)
            prev = startNode
        else:
            prev.next = startNode.next
        startNode = startNode.next
    return


def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

myList = [1]
printLinkedList(listToLinkedList(myList))
lList = listToLinkedList(myList)
removeDupes(lList)
printLinkedList(lList)

myList = [1, 2, 3, 1]
printLinkedList(listToLinkedList(myList))
lList = listToLinkedList(myList)
removeDupes(lList)
printLinkedList(lList)

myList = [1, 2, 3, 1, 4]
printLinkedList(listToLinkedList(myList))
lList = listToLinkedList(myList)
removeDupes(lList)
printLinkedList(lList)

myList = [1, 2, 3, 4, 1, 5, 6, 1, 7]
printLinkedList(listToLinkedList(myList))
lList = listToLinkedList(myList)
removeDupes(lList)
printLinkedList(lList)

```

```python
# 2.2 Return Kth to Last: Implment an algorithm to find the kth to last element of a singly linked list.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")


# let's assume that the linked list is at least k nodes long
def kthToLast(startNode: Node, k: int):
    firstPointer = startNode
    secondPointer = startNode

    for i in range(k - 1):
        secondPointer = secondPointer.next

    while secondPointer.next != None:
        firstPointer = firstPointer.next
        secondPointer = secondPointer.next

    return firstPointer.data

myList = [1, 2, 3, 4, 5, 6, 7]
myLinked = listToLinkedList(myList)
k = 3

assert kthToLast(myLinked, k) == 5, "test case 1 failed"
print("All 2.2 test cases passed")
```

```python
# 2.3 Delete Middle Node: Implement an alogrithm to delete a node in the middle (i.e. any node but the first and last node, not necessarily the exact middle) of a singly
# linked list, given only access to that node.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

def deleteMidNode(midNode: Node):
    while midNode.next != None and midNode.next.next != None:
        midNode.data = midNode.next.data
        midNode = midNode.next
    
    # erase the last node
    midNode.data = midNode.next.data
    midNode.next = None

myList = [1, 2, 3, 4, 5, 6, 7, 8, 9]
lList = listToLinkedList(myList)
deleteMidNode(lList.next.next.next.next)
printLinkedList(lList)
```

```python
# 2.4 Partition: Write code to partition a linked list around a value x, such that all nodes less than x come
# before all nodes greater than or equal to x. lf x is contained within the list, the values of x only need
# to be after the elements less than x (see below). The partition element x can appear anywhere in the
# "right partition"; it does not need to appear between the left and right partitions.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

def partition(headNode: Node, partitionVal: int):
    if headNode == None or headNode.next == None:
        return headNode

    leftListDummyHead = Node(0)
    actualLeftDummyHead = leftListDummyHead
    rightListDummyHead = Node(0)
    actualRightDummyHead = rightListDummyHead

    while headNode.next != None:
        if headNode.data >= partitionVal:
            rightListDummyHead.next = Node(headNode.data)
            rightListDummyHead = rightListDummyHead.next
        else:
            leftListDummyHead.next = Node(headNode.data)
            leftListDummyHead = leftListDummyHead.next
        headNode = headNode.next

    # merge
    leftListDummyHead.next = actualRightDummyHead.next

    return actualLeftDummyHead.next

myList = [1, 9, 8, 2, 5, 3, 4, 7, 6]
lList = listToLinkedList(myList)
solutionList = partition(lList, 5)
printLinkedList(solutionList)
```

```python
# 2.5 Sum Lists: You have two numbers represented by a linked list, where each node contains a single
# digit. The digits are stored in reverse order, such that the 1 's digit is at the head of the list. Write a
# function that adds the two numbers and returns the sum as a linked list.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

def sumLists(head1: Node, head2: Node):

    if head1 == None and head2 == None:
        return head1
    if head1 == None and head2 != None:
        return head2
    if head2 == None and head1 != None:
        return head1

    solutionHeadDummy = Node(0)
    actualHead = solutionHeadDummy
    carryOver = 0

    while head1 != None and head2 != None:
        sumValue = carryOver + head1.data + head2.data
        carryOver = sumValue // 10
        onesDigit = sumValue % 10
        solutionHeadDummy.next = Node(onesDigit)
        solutionHeadDummy = solutionHeadDummy.next
        head1 = head1.next
        head2 = head2.next
    
    if head1 != None:
        while head1 != None:
            sumValue = carryOver + head1.data
            carryOver = sumValue // 10
            onesDigit = sumValue % 10
            solutionHeadDummy.next = Node(onesDigit)
            solutionHeadDummy = solutionHeadDummy.next
            head1 = head1.next
        if carryOver:
            solutionHeadDummy.next = Node(carryOver)
            carryOver = 0
    
    if head2 != None:
        while head2 != None:
            sumValue = carryOver + head2.data
            carryOver = sumValue // 10
            onesDigit = sumValue % 10
            solutionHeadDummy.next = Node(onesDigit)
            solutionHeadDummy = solutionHeadDummy.next
            head2 = head2.next
        if carryOver:
            solutionHeadDummy.next = Node(carryOver)
            carryOver = 0
    
    if carryOver:
        solutionHeadDummy.next = Node(carryOver)
    
    return actualHead.next

num1 = [9, 9, 9]
num2 = [1]
linked1 = listToLinkedList(num1)
linked2 = listToLinkedList(num2)
solutionList = sumLists(linked1, linked2)
printLinkedList(solutionList)
```

```python
# 2.6 Palindrome: Implement a function to check if a linked list is a palindrome.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")


# 1 3 5 7

def isPalindrome(head: Node):
    tmpHead = head

    dummyHead = Node(0)

    while tmpHead != None:
        newNode = Node(tmpHead.data)
        newNode.next = dummyHead
        dummyHead = newNode
        tmpHead = tmpHead.next

    while head != None:
        if head.data != dummyHead.data:
            return False
        head = head.next
        dummyHead = dummyHead.next
    return True

num1 = [1, 5, 9, 5, 1]
linked1 = listToLinkedList(num1)

assert isPalindrome(linked1) == True, "test case 1 failed"
print("All 2.6 test cases passed")
```

```python
# 2.7 Intersection: Given two (singly) linked lists, determine if the two lists intersect. Return the inter-
# secting node. Note that the intersection is defined based on reference, not value. That is, if the kth
# node of the first linked list is the exact same node (by reference) as the jth node of the second
# linked list, then they are intersecting.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

def intersectingNode(head1: Node, head2: Node):
    if head1 == None or head1.next == None or head2 == None or head2.next == None:
        return (False, Node(0))
    
    tmpHead1 = head1
    length1 = 0
    while tmpHead1 != None:
        length1 += 1
        tmpHead1 = tmpHead1.next
    
    tmpHead2 = head2
    length2 = 0
    while tmpHead2 != None:
        length2 += 1
        tmpHead2 = tmpHead2.next
    
    lengthDiff = abs(length1 - length2)

    tmpHeadPointer1 = head1
    tmpHeadPointer2 = head2
    if length1 > length2:
        for i in range(lengthDiff):
            tmpHeadPointer1 = tmpHeadPointer1.next
    if length2 > length1:
        for i in range(lengthDiff):
            tmpHeadPointer2 = tmpHeadPointer2.next
    
    while tmpHeadPointer1 != None and tmpHeadPointer2 != None:
        if tmpHeadPointer1 == tmpHeadPointer2:
            return (True, tmpHeadPointer1)
        tmpHeadPointer1 = tmpHeadPointer1.next
        tmpHeadPointer2 = tmpHeadPointer2.next
    
    return (False, Node(0))


myList1 = [3, 1, 5, 6, 2, 7]
lList1 = listToLinkedList(myList1)

myList2 = [2, 4, 3, 1, 5]
lList2 = listToLinkedList(myList2)

lList2.next.next = lList1.next.next.next

printLinkedList(lList1)
printLinkedList(lList2)

assert intersectingNode(lList1, lList2) == (True, lList1.next.next.next), "test case 1 failed"
print("All 2.7 tests passed")
```

```python
# 2.8 Loop Detection: Given a circular linked list, implement an algorithm that returns the node at the
# beginning of the loop.
# DEFINITION
# Circular linked list: A (corrupt) linked list in which a node's next pointer points to an earlier node, so
# as to make a loop in the linked list.

from typing import List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def listToLinkedList(arr: List[int]):
    dummyNode = Node(0)
    head = dummyNode
    for i in range(len(arr)):
        dummyNode.next = Node(arr[i])
        dummyNode = dummyNode.next
    return head.next

def printLinkedList(startNode: Node):
    while(startNode != None):
        print(str(startNode.data) + " ", end=" ")
        startNode = startNode.next
    print("")

# assume minimum number of nodes is 2
def getCircleHead(head: Node):
    slowPtr, fastPtr = head, head

    slowPtr = slowPtr.next
    fastPtr = fastPtr.next.next

    while slowPtr != fastPtr:
        slowPtr = slowPtr.next
        fastPtr = fastPtr.next.next
    
    slowPtr = head

    while (slowPtr != fastPtr):
        slowPtr = slowPtr.next
        fastPtr = fastPtr.next
    
    return slowPtr

myList1 = [3, 1, 5, 6, 2, 7]
lList1 = listToLinkedList(myList1)

lList1.next.next.next.next.next = lList1.next.next

assert getCircleHead(lList1).data == 5, "test case 1 failed"
print("All 2.8 tests passed")
```
