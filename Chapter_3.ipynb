```python
# 3.1 Three In One: Describe how you could use a single array to implement three stacks.

# Start with an array of fixed size, keep pointers / indices at the start of three equal partitions of the array,
# along with the size of each array. Allow for all the normal operations on a stack where the implementation uses
# an array (push adds to the first non-empty index at the end of the subarray, pop removes the element at the end
# of the array, peek only allows looking at the end of the array). If any push would cause a subarray to exceed max
# capacity, create a new array with double the space, create new partition sizes and indices, and copy over all of
# the data. Could also implement compaction logic that would resize the array smaller if it becomes very large and then
# a lot of data is popped out of the stacks

# 3.2 Stack Min: How would you design a stack which, in addition to push and pop, has a function min which returns the
# minimum element? Push, pop, and min should all operate in O(1) time.

# Keep a member of the stack class called minStack which tracks the minimum element. When a push operation occurs,
# only push it to minStack if its lower than or equal to the top value of minStack. When a pop operation occurs,
# if the value equals the top value of minStack, then pop it from minStack as well. The top value of minStack
# will always have the minimum value in the stack

# 3.3 Stack of Plates: Imagine a (literal) stack of plates. If the stack gets too high, it might topple.
# Therefore, in real life, we would likely start a new stack when the previous stack exceeds some
# threshold. Implement a data structure SetOfStacks that mimics this. SetOfStacks should be
# composed of several stacks and should create a new stack once the previous one exceeds capacity.
# SetOfStacks. push () and SetOfStacks. pop () should behave identically to a single stack
# (that is, pop ( ) should return the same values as it would if there were just a single stack).

# FOLLOW UP
# Implement a function popAt (int index) which performs a pop operation on a specific sub-stack.

class SetOfStacks:
    def __init__(self, maxSubSize: int):
        self.stacks = []
        self.maxSubSize = maxSubSize
    
    def push(self, value: int):
        if len(self.stacks) == 0:
            self.stacks.append([value])
        else:
            valueAdded = False
            for i in range(len(self.stacks)):
                if len(self.stacks[i]) < self.maxSubSize:
                    self.stacks[i].append(value)
                    valueAdded = True
                    break
            if not valueAdded: # all the current sub stacks were full, create a new one
                self.stacks.append([value])
    
    def pop(self):
        for i in range(len(self.stacks) - 1, -1, -1):
            if len(self.stacks[i]) > 0:
                self.stacks[i].pop()
                break
    
    def popAt(self, index: int):
        if index < 0 or index >= len(self.stacks):
            return
        if len(self.stacks[index]) == 0:
            return
        else:
            self.stacks[index].pop()
    
    def printStack(self):
        print(self.stacks)


myStacks = SetOfStacks(3)

for i in range(11):
    myStacks.push(i)


myStacks.printStack()
myStacks.pop()
myStacks.printStack()
myStacks.pop()
myStacks.printStack()
myStacks.pop()
myStacks.printStack()

myStacks.popAt(1)
myStacks.printStack()

print("All 3.3 test cases pass")
```
